```python
#Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.
#Ans-

import pandas as pd

def reindex_dataframe(df):
    new_index = pd.Index(range(1, 2*len(df)+1, 2))
    df = df.set_index(new_index)
    return df

'''Here, we first create a new index by using the range function, which starts from 1 and increments by 2 for each row. 
We then set this new index as the index of the DataFrame using the set_index function. Finally, we return the updated DataFrame.

To use this function with a sample DataFrame df with columns 'A', 'B', and 'C', we can do the following:'''

import pandas as pd

# create a sample DataFrame
df = pd.DataFrame({'A': [10, 20, 30, 40], 'B': [1, 2, 3, 4], 'C': ['X', 'Y', 'Z', 'W']})

# call the reindex_dataframe function
df = reindex_dataframe(df)

# print the updated DataFrame
print(df)
```

```python
#Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. 
#The function should print the sum to the console.
#For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should calculate and print the sum of the first three values, which is 60.

#Ans-

df2 = pd.DataFrame({'Name' : ['a', 'b', 'c', 'd', 'e'], 
                   'Values' : [10, 20, 30, 40, 50]})
df2
```

```python
def sum_first_three(df2):
    sum = 0
    for index, row in df2.iterrows():
        if index < 3:
            sum += row['Values']
        else:
            break
    print("Sum of first three values in 'Values' column: ", sum)
```

```python
#Q5. How are DataFrame.size() and DataFrame.shape() different?
#Ans-

'''DataFrame.size() and DataFrame.shape() are both methods in Pandas that can be used to obtain information about a DataFrame, but they provide different types of information.

DataFrame.size() returns the total number of elements in a DataFrame, which is equal to the number of rows multiplied by the number of columns. It does not provide information about the dimensions of the DataFrame, only the total number of elements. 
For example, if a DataFrame has 3 rows and 4 columns, DataFrame.size() will return 12.

On the other hand, DataFrame.shape() returns a tuple that contains the dimensions of the DataFrame, in the form (number of rows, number of columns). For example, if a DataFrame has 3 rows and 4 columns, DataFrame.shape() will return (3, 4). 
It does not provide information about the total number of elements in the DataFrame.

Here is an example to illustrate the difference:'''

import pandas as pd

data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df5 = pd.DataFrame(data)

print("DataFrame size:", df5.size)      # Output: 9
print("DataFrame shape:", df5.shape)    # Output: (3, 3)
```

```python
#Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. 
#Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.

#Ans-

'''To extract the username part of each email address in the 'Email' column of the Pandas DataFrame df and create a new column 'Username' with this information, you can use the split() method to split each email address at the '@' symbol and then select the first element of the resulting list. 
Here's an example Python function that does this:'''

def extract_username(df):
    # Split the email addresses at the '@' symbol and select the first element of the resulting list
    df['Username'] = df['Email'].apply(lambda x: x.split('@')[0])
    return df

'''This function creates a new column 'Username' in the original DataFrame df and assigns to it the results of the apply() method applied to the 'Email' column. 
The lambda function passed to apply() splits each email address at the '@' symbol and selects the first element of the resulting list, which is the username. 
The resulting DataFrame contains both the original 'Email' column and the new 'Username' column.'''
```

```python
#Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. 
#The function should return a new DataFrame that contains only the selected rows.

#Ans-

import pandas as pd

# create a sample DataFrame
df = pd.DataFrame({'A': [1, 20, 3, 4, 7, 8, 12, 13], 'B': [1, 20, 3, 4, 50, 6, 7, 8], 'C': [10, 12, 3, 8, 9, 45, 13, 17]})
df
```

```python
#Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.
#Ans-

'''Here's a Python function that takes a Pandas DataFrame df and calculates the mean, median, and standard deviation of the values in the 'Values' column using built-in Pandas functions mean(), median(), and std() respectively:'''

import pandas as pd

def calculate_stats(df):
    mean = df['Values'].mean()
    median = df['Values'].median()
    std_dev = df['Values'].std()
    return mean, median, std_dev

df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
mean, median, std_dev = calculate_stats(df)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```

```python
#Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. 
#The moving average should be calculated using a window of size 7 and should include the current day.

#Ans-

'''To create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days, we can use the rolling() function provided by pandas. 
Here's an example Python function that implements this:'''

import pandas as pd

def calculate_moving_average(df):
    # sort the DataFrame by date
    df = df.sort_values('Date')

    # calculate the moving average
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df
```

```python
#Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

#Ans-

import pandas as pd

def select_rows_between_dates(df):
    # Convert 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Select rows between '2023-01-01' and '2023-01-31'
    mask = (df['Date'] >= '2023-01-01') & (df['Date'] <= '2023-01-31')
    selected_rows = df.loc[mask]
    
    return selected_rows
```
